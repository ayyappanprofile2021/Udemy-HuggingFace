```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
```

```python
model_name = "facebook/bart-large-cnn"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
text = """Artificial Intelligence (AI) has become one of the most transformative technologies of the 21st century. It is being used across industries such as healthcare, finance, education, manufacturing, transportation, and entertainment. AI systems can analyze large amounts of data, identify patterns, and make predictions much faster than humans. In healthcare, AI helps doctors diagnose diseases, recommend treatments, and monitor patients remotely. Financial institutions use AI to detect fraudulent transactions, assess credit risks, and automate customer service through chatbots."""
```

```python
#Tokenize the input text.
inputs = tokenizer.encode("summarize: " + text, return_tensors = "pt", max_length=512, truncate=True)
```

```python
#Generate the summary
summary_ids = model.generate(inputs, max_length=50, min_length=25, length_penalty=2.0, num_beams=4)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print(summary)
```

```python
text = "Artificial Intelligence is amazing."

token_ids = tokenizer.encode(text)

tokens = tokenizer.convert_ids_to_tokens(token_ids)

print(tokens)
```
